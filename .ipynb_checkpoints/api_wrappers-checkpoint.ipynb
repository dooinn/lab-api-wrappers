```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python

your_client_id = "YOUR CLIENT ID"
your_client_secret = "YOUR CLIENT SECRET"

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=your_client_id,
                                                           client_secret=your_client_secret))

results = sp.search(q='artist:Los Hermanos', limit=5, type='track')
results
```

```python
results['tracks']['items']
```

```python
def get_playlist_tracks(username, playlist_id):

    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']

    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])

    return tracks

tracks = get_playlist_tracks("spotify", "12Wbv8sIx84T5uh6iOoJ7V")
tracks
```

```python
track_ids = [track['track']['uri'] for track in tracks]
track_ids
```

```python
def get_artists_and_ids_from_playlist(playlist_id):

    tracks_from_playlist = get_playlist_tracks("spotify", playlist_id)

    artists = []
    artists_ids = []

    for track in tracks_from_playlist:
        artists_info = track['track']['artists']

        for artist_info in artists_info:
            artists.append(artist_info['name'])
            artists_ids.append(artist_info['id'])

    return list(set(artists)), list(set(artists_ids))

artists, artists_ids = get_artists_and_ids_from_playlist("12Wbv8sIx84T5uh6iOoJ7V")
artists, artists_ids
```
